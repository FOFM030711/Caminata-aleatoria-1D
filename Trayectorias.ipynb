Hagamos $X_0=K=20$

$n= 1,000$ pasos

$p=0.5$


```python
 # liberias
import random  # Importa la librería random para generar números aleatorios
import matplotlib.pyplot as plt  # Importa matplotlib para graficar
```

```python
def caminata_aleatoria(n_pasos=1000, p=0.5):
    # Define la función de la caminata aleatoria
    # n_pasos es el número de movimientos
    # p es la probabilidad de moverse a la derecha

    if p < 0 or p > 1:
        # Verifica que la probabilidad esté en el rango válido
        raise ValueError("p debe estar entre 0 y 1")
        # Lanza un error si p no es válida

    posiciones = [20]
    # Crea una lista llamada posiciones
    # Inicia en 0 (posición inicial de la partícula)

    for i in range(n_pasos):
        # Repite el proceso n_pasos veces

        r = random.random()
        # Genera un número aleatorio uniforme entre 0 y 1

        if r < p:
            # Si el número aleatorio es menor que p
            nuevo = posiciones[-1] + 1
            # Se mueve una unidad a la derecha
        else:
            nuevo = posiciones[-1] - 1
            # En caso contrario se mueve una unidad a la izquierda

        posiciones.append(nuevo)
        # Agrega la nueva posición a la lista

    return posiciones
    # Devuelve la lista completa de posiciones
```

```python
# Parámetros
n = 1000
# Número total de pasos que dará la partícula

p = 0.5
# Probabilidad de moverse a la derecha

trayectoria = caminata_aleatoria(n, p)
# Ejecuta la función y guarda la trayectoria generada

plt.figure()
# Crea una nueva figura para la gráfica

plt.plot(trayectoria)
# Dibuja la trayectoria como una sola línea

plt.title(f"Caminata Aleatoria 1D (p = {p})")
# Coloca el título de la gráfica

plt.xlabel("Paso")
# Etiqueta del eje horizontal (número de paso)

plt.ylabel("Posición")
# Etiqueta del eje vertical (posición en la recta)

plt.axhline(0)
# Dibuja una línea horizontal en y=0 para marcar el origen
plt.grid(True)
# Activa la cuadrícula
plt.show()
# Muestra la gráfica en pantalla
```

Simulación de $100$ juegos para ver cual es la probabilidad de que obtengamos cero

¿Cuál es la probabilidad de que la trayectoria llegue a cero?

```python
all_trajectories = []
# Inicializa una lista vacía para almacenar todas las trayectorias

for _ in range(100):
    # Bucle que se repite 100 veces para generar 100 simulaciones
    trayectoria_individual = caminata_aleatoria(n_pasos=1000, p=0.5)
    # Llama a la función caminata_aleatoria para generar una trayectoria
    # con 1000 pasos y probabilidad p=0.5
    all_trajectories.append(trayectoria_individual)
    # Agrega la trayectoria generada a la lista all_trajectories

print(f"Se han generado {len(all_trajectories)} trayectorias.")
# Imprime la cantidad de trayectorias generadas para verificar
```

## Verificar Alcance de Cero

Para cada una de las 100 trayectorias simuladas, verificar si la posición 0 está presente en algún momento de la trayectoria.


```python
reached_zero = []
# Inicializa una lista vacía para almacenar si cada trayectoria alcanzó cero

for trayectoria in all_trajectories:
    # Itera sobre cada trayectoria generada
    if 0 in trayectoria:
        # Verifica si la posición 0 está presente en la trayectoria actual
        reached_zero.append(True)
        # Si 0 está en la trayectoria, agrega True a la lista
    else:
        reached_zero.append(False)
        # Si 0 no está en la trayectoria, agrega False a la lista

# Calcula el número de trayectorias que alcanzaron cero
num_reached_zero = reached_zero.count(True)

print(f"Número total de trayectorias: {len(all_trajectories)}")
print(f"Número de trayectorias que alcanzaron el cero: {num_reached_zero}")
```


El paso anterior identificó correctamente cuántas trayectorias llegaron a cero. Ahora, para cumplir con el requisito de la tarea de calcular e imprimir la probabilidad, necesito dividir el número de trayectorias que llegaron a cero entre el número total de trayectorias.



```python
probabilidad_cero = num_reached_zero / len(all_trajectories)

print(f"La probabilidad de que una trayectoria alcance el cero es: {probabilidad_cero:.2f}")
```
