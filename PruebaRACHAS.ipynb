A continuacion, consideremos la siguiente hipotesis nula:


$H_0$:  los numeros de la muestra ${x_i}$ son independientes

y luego

$H_1$:  los numeros de la muestra ${x_i}$ no son independientes

```python
import numpy as np
from scipy.stats import norm
```

```python
matriz
```

```python

# Convertimos la matriz en vector
datos = matriz.flatten()
n = len(datos)
```

```python

print("Cantidad total de datos (n):", n)
```

<span style="color:orange;"></span>**Paso 1**: Determine la secuencia $S$ de $0$ y $1$ la + y -, así:



*   Si $x_i  \leq x_{i-1}$n escribir $0$ (o "-")
*   Si $x_i  > x_{i-1}$ escribir $0$ (o "+")



```python
# Construcción de la secuencia S


secuencia = []

for i in range(1, n):
    if datos[i] <= datos[i-1]:
        secuencia.append(0)
    else:
        secuencia.append(1)


```

```python
print("\nSecuencia generada:")
print(secuencia)
```

La secuencia contiene $n-1$ numeros



<span style="color:orange;"></span>**Paso 2**: Contar el numero de rachas observadas $c_o$
Una racha viene dada por el numero de ceros o unas secuencias que la forman.


```python

rachas = 1

for i in range(1, len(secuencia)):
    if secuencia[i] != secuencia[i-1]:
        rachas += 1

print("\nNúmero de rachas observadas (c_o):", rachas)
```


<span style="color:brown;"></span>**Paso 3**: Calcule el valor esperado y la varianza del numero de rachas, asi como el estadistico $Z_0$ mediante:



```python

mu = (2*n - 1) / 3
sigma = (16*n - 29) / 90   # ← SIN RAÍZ

Z0 = (rachas - mu) / sigma


```


$\mu_{c_0} = \frac{2n - 1}{3}$


```python
print("\nValor esperado (μ):", round(mu,4))
```


$\sigma_{c_0}$=$\frac{16n-29}{90}$



```python

print("Sigma (σ):", round(sigma,4))
```

$Z_0$=$\frac{c_o-\mu}{\sigma}$


Dado que $C_o$ sigue una distribución normal de media $μ_{c_o}$ y varianza $σ^2_{C_o}$ y $Z_0$ una distribución normal (0,1). El intervalo de aceptación:
$$
-≤Z_0\le Z_{\frac{α}{2}} $$
o bien
$$|Z_0| ≤Z_{\frac{α}{2}}$$

```python

print("Estadístico Z0:", round(Z0,4))
```

```python

alpha = 0.05
Z_critico = norm.ppf(1 - alpha/2)

```

```python

print("\nValor crítico ±Zα/2:", round(Z_critico,4))

print("\n================ RESULTADO FINAL ================")

if abs(Z0) < Z_critico:
    print("NO se rechaza H0")
    print("Conclusión: La secuencia PUEDE considerarse ALEATORIA.")
else:
    print("Se rechaza H0")
    print("Conclusión: La secuencia NO es aleatoria.")

print("=================================================")
```

<span style="color:blue;"></span>**Interpretación:**

El estadístico $Z$ calculado (-0.0827) cae dentro del intervalo crítico $[-1.96, 1.96]$, lo que indica que no hay evidencia suficiente para rechazar la hipótesis de aleatoriedad.

<span style="color:purple;">Por lo tanto no se rechaza la hipotesis
$H_0:$ Los números de la muestra $\{x_i\}$ son independientes(aleatorios)</span>
