```python


x, y, z = symbols('x y z')

M_sympy = Matrix(M)

pi_vector = Matrix([[x, y, z]])


eq1 = pi_vector * M_sympy - pi_vector


eq_system = [
    eq1[0],
    eq1[1],
    x + y + z - 1
]

solution = solve(eq_system, (x, y, z))

print(f"la solucion es  [x, y, z] is: {solution}")
```
