```python
import random as r
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Beneficio(ventas,compras=400):
    if ventas<compras:
        beneficio=ventas*1000+(compras-ventas)*250-compras*750
    else:
        beneficio=compras*1000-compras*750
    return beneficio
```

```python
def Venta():
  V=r.random()
  if 0 <= V < 0.22:
     U_v= 100
  elif 0.22 <= V <0.47:
     U_v=200
  elif 0.47<= V < 0.58:
     U_v=300
  elif 0.58 <= V < 0.83:
     U_v=400
  elif 0.83 <= V < 0.94:
     U_v=500
  else:
     U_v=600
  return U_v
```

```python
Venta()
```

```python
def experimento(n):
    promedio = []
    prom = 0
    for I in range(n):
        v = Venta()
        b = Beneficio(v)
        prom = (I * prom + b) / (I + 1)
        promedio.append(prom)
    return promedio
```

```python
L=experimento(850)
```

```python
plt.plot(L)
```

```python
np.mean(L)
#despues de La gráfica
```

Paso 8. Que el experimento anterior fijamos el tamaño de la muestra del experimento de simulación

```python
def simulación(n):
    lista=[]
    for i in range(n): #bucle
        e=np.mean(experimento(850))
        lista.append(e)
    return lista
```

```python
s= simulación(50)
```
