#<span style="color:brown;">**CAMITAS ALEATORIAS 1D**</span>
<span style="color:blue;"></span>




Una **caminata aleatoria unidimensional** es un proceso estocástico que describe una secuencia de pasos aleatorios en una línea. En su forma más simple, un "caminante" comienza en una posición inicial (comúnmente $x=0$) y en cada paso se mueve una unidad a la derecha o una unidad a la izquierda.

*   **$p$**: Es la probabilidad de que el caminante se mueva un paso a la **derecha**.
*   **$q$**: Es la probabilidad de que el caminante se mueva un paso a la **izquierda**.

La suma de estas probabilidades debe ser 1, ya que en cada paso el caminante solo puede ir a la derecha o a la izquierda:

$$p + q = 1$$

#<span style="color:orange;">*Posición del Caminante *</span>
<span style="color:blue;"></span>



Sea $S_n$ la posición del caminante después de $n$ pasos. Si $X_i$ es una variable aleatoria que representa el $i$-ésimo paso ($+1$ si va a la derecha, $-1$ si va a la izquierda), entonces:

$$S_n = \sum_{i=1}^{n} X_i$$

Donde $P(X_i = 1) = p$ y $P(X_i = -1) = q$.



#<span style="color:orange;">*Valor Esperado de la Posición *</span>
<span style="color:blue;"></span>



El valor esperado de la posición del caminante después de $n$ pasos se calcula como:

$$\text{E}[S_n] = n(p - q)$$

Si $p = q = 0.5$ (caminata aleatoria simétrica), entonces $\text{E}[S_n] = 0$.

#<span style="color:orange;">*Varianza de la Posición*</span>
<span style="color:blue;"></span>



La varianza de la posición del caminante después de $n$ pasos se define como:

$$\text{Var}(S_n) = n(1 - (p-q)^2) = 4npq$$

Para una caminata aleatoria simétrica ($p=q=0.5$), la varianza es


$\text{Var}(S_n) = n(1 - 0^2) = n$ o $4n(0.5)(0.5) = n$

```python
 # liberias
import random  # Importa la librería random para generar números aleatorios
import matplotlib.pyplot as plt  # Importa matplotlib para graficar
```

```python
def caminata_aleatoria(n_pasos=100, p=0.5):
    # Define la función de la caminata aleatoria
    # n_pasos es el número de movimientos
    # p es la probabilidad de moverse a la derecha

    if p < 0 or p > 1:
        # Verifica que la probabilidad esté en el rango válido
        raise ValueError("p debe estar entre 0 y 1")
        # Lanza un error si p no es válida

    posiciones = [0]
    # Crea una lista llamada posiciones
    # Inicia en 0 (posición inicial de la partícula)

    for i in range(n_pasos):
        # Repite el proceso n_pasos veces

        r = random.random()
        # Genera un número aleatorio uniforme entre 0 y 1

        if r < p:
            # Si el número aleatorio es menor que p
            nuevo = posiciones[-1] + 1
            # Se mueve una unidad a la derecha
        else:
            nuevo = posiciones[-1] - 1
            # En caso contrario se mueve una unidad a la izquierda

        posiciones.append(nuevo)
        # Agrega la nueva posición a la lista

    return posiciones
    # Devuelve la lista completa de posiciones
```

```python
# Parámetros
n = 100
# Número total de pasos que dará la partícula

p = 0.9
# Probabilidad de moverse a la derecha (90%)

trayectoria = caminata_aleatoria(n, p)
# Ejecuta la función y guarda la trayectoria generada

plt.figure()
# Crea una nueva figura para la gráfica

plt.plot(trayectoria)
# Dibuja la trayectoria como una sola línea

plt.title(f"Caminata Aleatoria 1D (p = {p})")
# Coloca el título de la gráfica

plt.xlabel("Paso")
# Etiqueta del eje horizontal (número de paso)

plt.ylabel("Posición")
# Etiqueta del eje vertical (posición en la recta)

plt.axhline(0)
# Dibuja una línea horizontal en y=0 para marcar el origen
plt.grid(True)
# Activa la cuadrícula
plt.show()
# Muestra la gráfica en pantalla
```

##¿Qué preguntas surgen?

*   ¿Cuánto tiempo tarda en regresar al origen?

```python
# Tiempo del primer regreso al origen
tiempo_regreso = None

for i in range(1, len(trayectoria)):
    if trayectoria[i] == 0:
        tiempo_regreso = i
        break

if tiempo_regreso is not None:
    print("Primer regreso al origen en el paso:", tiempo_regreso)
else:
    print("No regresó al origen")
```

*   ¿Que sucede cuando la probabilidad es igual a 1 $(p=1)$?

```python
# Parámetros
n = 100
# Número total de pasos que dará la partícula

p = 1
# Probabilidad de moverse a la derecha (60%)

trayectoria = caminata_aleatoria(n, p)
# Ejecuta la función y guarda la trayectoria generada

plt.figure()
# Crea una nueva figura para la gráfica

plt.plot(trayectoria)
# Dibuja la trayectoria como una sola línea

plt.title(f"Caminata Aleatoria 1D (p = {p})")
# Coloca el título de la gráfica

plt.xlabel("Paso")
# Etiqueta del eje horizontal (número de paso)

plt.ylabel("Posición")
# Etiqueta del eje vertical (posición en la recta)

plt.axhline(0)
# Dibuja una línea horizontal en y=0 para marcar el origen
plt.grid(True)
# Activa la cuadrícula
plt.show()
# Muestra la gráfica en pantalla
```


Como podeos ver en la imagen $(p = 1)$, la partícula se mueve siempre hacia la derecha,
por lo que cada paso es $+1$. En consecuencia, la posición después
de $n$ pasos es

$
X_n = n.
$

En este caso, la caminata deja de ser aleatoria y se convierte en
un movimiento determinista con crecimiento lineal. Además, la varianza es

$$
\mathrm{Var}(X_n) = 4np(1-p) = 0,
$$

lo que indica que no existe variabilidad en la trayectoria.






*   ¿Cuál fue el máximo alejamiento?



```python
# Máximo alejamiento respecto al origen
max_alejamiento = max(abs(x) for x in trayectoria)

print("Máximo alejamiento:", max_alejamiento)
```

*   ¿Qué sucede si la probabilidad sea mayor a 1 $(p>1)$?


Si $p > 1$, el modelo pierde sentido, ya que una probabilidad no puede
ser mayor que $1$. Tengamos presente que por definición, las probabilidades solo pueden tomar valores entre $0$ y $1$.

Por lo que una caminata aleatoria únicamente es válida cuando
$p \in [0,1]$; fuera de ese intervalo, el proceso deja de tener
interpretación probabilística (visto en clase).

*   ¿Cuánto se aleja en promedio?

```python
# Distancia promedio respecto al origen
distancia_promedio = sum(abs(x) for x in trayectoria) / len(trayectoria)

print("Alejamiento promedio:", distancia_promedio)
```

*   ¿Cuál es la posición promedio?

```python
# Posición promedio
posicion_promedio = sum(trayectoria) / len(trayectoria)

print("Posición promedio:", posicion_promedio)
```

*   ¿Regresa y cuántas veces regresa al origen?

```python
# Número de veces que regresa al origen (sin contar el inicio)
regresos = sum(1 for x in trayectoria[1:] if x == 0)

if regresos > 0:
    print("Sí regresó al origen.")
    print("Número de regresos:", regresos)
else:
    print("No regresó al origen.")
```

# Análisis y Conclusiones
Cuando $p = 0.5$, la caminata aleatoria no tiene una dirección preferente;
el movimiento es equilibrado y no existe una tendencia definida. Sin embargo,
debido a la naturaleza aleatoria del proceso, una trayectoria individual
puede alejarse considerablemente del origen.

Si $p > 0.5$, aparece una tendencia creciente que coincide con el valor esperado

$$
\mathbb{E}[X_n] = n(2p - 1).
$$

Por el contrario, si $p < 0.5$, la tendencia es decreciente, con el mismo
comportamiento en magnitud, pero en sentido opuesto.

En conjunto, aunque cada caminata es impredecible, su comportamiento promedio
sigue leyes estadísticas claras. Este modelo permite comprender cómo pequeñas
variaciones en la probabilidad pueden generar diferencias significativas con
el paso del tiempo, siendo útil para describir fenómenos como el movimiento
browniano discreto y procesos financieros simplificados.
