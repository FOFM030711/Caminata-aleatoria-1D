**METODO DE UNIFORMIZACION**

Dado ${X(t), t\geq0}$ queremos calcular la pmf de $X(t)$.
Supongamos conocida $X(0)$. Queremos:
* $$P(X(t)=j)=\sum_{i=1}^{n} P(X(t)= j  \lvert  X(0)=i)P(x(0))= i,     1 \leq j \leq N $$

Entonces, necesitamos calcular $$ P(X(t)= j  \lvert  X(0)=i)=p_{ij}(t)$$
* $P(t)=\[p_{ij}(t)]\$


y recordar que $R=\[r_{ij}]\$

Sea $r \geq max \{r_i\},  1 \leq j \leq N$



Definimos $ \hat{P}=  [ \hat{p}_{ij}]\ $ así:



**Teorema:**La matriz de probabilidades de transición
$ P(t)= [p_{ij}(t)]\ $ esta dada por:
$$ P(t)= \sum_{k=1}^{∞} e ^{-rt} \frac{(rt)^k}{k!}  \hat{P}^k$$

Este **Teorema** nops permite aproximar $P(t)$ usando $M$ términos de la serie. ¿Cómo elegir $M$?
$$ M\approx max \{ rt+5* \sqrt{rt}, 20 \} $$

**Ejercicio:**



1.   Escribir $P(t)$ usando el teorema anterior
2.   Calcular $P(0.5), P(1)$ y $P(5)$
3.   Verificar que $P(1)=P(0.5)P(0.5)$ según las ecuaciones de **Chapman-Kolmogorov**





```python
import sympy as sp
from sympy import Matrix, symbols
```

```python


# matriz de R
R=sp.Matrix([[0, 2,3,0], [4,0,2,0] , [0,2,0,2] , [1,0,3,0]])
R
```


Escribir $P(t)$ usando el teorema anterior

```python
r_i = [sum(row) for row in R.tolist()]
r = max(r_i)
print(f"Individual rates (r_i): {r_i}")
print(f"Maximum rate (r): {r}")

n_states = R.shape[0]
I = sp.eye(n_states)  # Matriz identidad del tamaño del número de estados

P_hat = sp.zeros(n_states, n_states)
for i in range(n_states):
    for j in range(n_states):
        if i == j:
            # Calcula la probabilidad de permanecer en el mismo estado
            P_hat[i, j] = 1 - r_i[i] / r
        else:
            # Calcula la probabilidad de transición entre estados distintos
            P_hat[i, j] = R[i, j] / r

print("Matriz P_hat:")
display(P_hat)
```

 Calcular $P(0.5), P(1)$ y $P(5)$

```python
def calculate_P_t(t_val, R_matrix):
    # Re-calculate r and P_hat within the function to ensure scope and reusability
    r_i_local = [sum(row) for row in R_matrix.tolist()]
    r_local = max(r_i_local)

    n_states_local = R_matrix.shape[0]
    P_hat_local = sp.zeros(n_states_local, n_states_local)
    for i in range(n_states_local):
        for j in range(n_states_local):
            if i == j:
                P_hat_local[i, j] = 1 - r_i_local[i] / r_local
            else:
                P_hat_local[i, j] = R_matrix[i, j] / r_local

    # Determine M based on the formula
    M = max(int(r_local * t_val + 5 * sp.sqrt(r_local * t_val)), 20)

    P_t = sp.zeros(n_states_local, n_states_local) # Initialize P(t) as a zero matrix
    current_P_hat_power = sp.eye(n_states_local) # P_hat^0 = Identity matrix

    for k in range(M):
        term = sp.exp(-r_local * t_val) * ((r_local * t_val)**k / sp.factorial(k)) * current_P_hat_power
        P_t += term
        current_P_hat_power *= P_hat_local # Calculate P_hat^(k+1) for the next iteration

    return P_t
```

```python
# Tiempos a evaluar
t_vals = [0.5, 1, 5]

# Resultados
P_results = {}

for t in t_vals:

    # Obtener P(t)
    P_t = calculate_P_t(t, R)

    # Guardar P(t)
    P_results[t] = P_t

    # Mostrar P(t)
    print(f"P({t}):")
    display(P_t)

    print("\n" + "-"*30 + "\n")
```

### Verificación de las Ecuaciones de Chapman-Kolmogorov

Vamos a verificar si $P(1) = P(0.5)P(0.5)$.

```python
# Obtener las matrices de transición previamente calculadas para t = 0.5 y t = 1
P_0_5 = P_results[0.5]
P_1 = P_results[1]

# Calcular P(0.5)² multiplicando la matriz P(0.5) por sí misma
P_0_5_squared = P_0_5 * P_0_5

# Mostrar el resultado de P(0.5)²
print("P(0.5) * P(0.5):")
display(P_0_5_squared)

# Mostrar la matriz P(1) obtenida directamente
print("\nP(1):")
display(P_1)

# Calcular la diferencia entre P(1) y P(0.5)²
# Si la ecuación de Chapman-Kolmogorov se cumple, esta diferencia debería ser cercana a cero
difference = P_1 - P_0_5_squared

# Mostrar la matriz de diferencias
print("\nDiferencia (P(1) - P(0.5)*P(0.5)):")
display(difference)

# Calcular la suma de los valores absolutos de todos los elementos de la diferencia
# Esto proporciona una medida global del error numérico
absolute_difference_sum = sum(abs(element) for element in difference)

# Mostrar el valor total del error
print(f"\nSuma de las diferencias absolutas: {absolute_difference_sum}")

# Verificar si el error es suficientemente pequeño
# Se utiliza una tolerancia de 10⁻⁹ para considerar errores de redondeo
if absolute_difference_sum < 1e-9:
    print("\nLa ecuación de Chapman-Kolmogorov P(1) = P(0.5)P(0.5) se verifica (dentro de una pequeña tolerancia).")
else:
    print("\nLa ecuación de Chapman-Kolmogorov P(1) = P(0.5)P(0.5) NO se verifica (o la tolerancia es demasiado pequeña).")
```
