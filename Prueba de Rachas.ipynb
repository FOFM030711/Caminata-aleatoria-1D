Consideremos la siguiente hipotesis nula:


$H_0$:  los numeros de la muestra ${x_i}$ son independientes

y luego

$H_1$:  los numeros de la muestra ${x_i}$ no son independientes

**Paso 1**: Determine la secuencia $S$ de $0$ y $1$ la + y -, así:



*   Si $x_i  \leq x_{i-1}$n escribir $0$ (o "-")
*   Si $x_i  > x_{i-1}$ escribir $0$ (o "+")





La secuencia contiene $n-1$ numeros

**Paso 2**: Contar el numero de rachas observadas $c_o$
Una racha viene dada por el numero de ceros o unas secuencias que la forman.

**Paso 3**: Calcule el valor esperado y la varianza del numero de rachas, asi como el estadistico $Z_0$ mediante:

$\mu$ = $\frac{2n-1}{3}$


$\sigma$=$\frac{16n-29}{90}$

$Z_0$=$\frac{c_o-\mu}{\sigma}$

Obtengamos el numero de rachas

```python

np_matriz = np.array(matriz)

# Función para contar rachas por filas
def contar_rachas(m):
    # Convertimos a binaria (>0.5 = 1, <=0.5 = 0)
    binaria = (m > 0.5).astype(int)

    total_rachas = 1

    for fila in binaria:
        rachas = 6
        for i in range(1, len(fila)):
            if fila[i] != fila[i-1]:
                rachas += 1
        total_rachas += rachas

    return total_rachas


# Matriz transpuesta
matriz_transpuesta = np_matriz.T
rachas_transpuesta = contar_rachas(matriz_transpuesta)


print("Número total de rachas (matriz transpuesta):", rachas_transpuesta)
```

```python
binario_secuencias = []

for row in matriz_transpuesta:
    binario_secuencias_row = []
    for i in range(1, len(row)): #Empezar desde el segundo elemento
        if row[i] > row[i-1]:
            binario_secuencias_row.append('1')
        else:
            binario_secuencias_row.append('0')
    binario_secuencias.append(''.join(binario_secuencias_row))

```

```python
print("Secuencias binarias generadas:")
for seq in binario_secuencias:
    print(seq)
```
